**Problem 1**: Prove that $$\lim_{x\to \infty} \frac{2+3x+x^{2}}{2x^{2}-3}=\frac{1}{2}$$


_Provide your answer here_ : 

Were I you, I would start this as 
\begin{align}
\left|\frac{2+3x+x^{2}}{2x^{2}-3}-\frac{1}{2} \right| = & \frac{1}{2}\left|\frac{1+3/x+2/x^{2}}{1-3/(2x^{2})}-1 \right|\\
= & \frac{1}{2|x|}\left|\frac{3+7/(2x)}{1-3/(2x^{2})}\right|
\end{align}
Then I would make sure that $x\gt\sqrt{3/2}$.  Taking $x\geq 2$ should suffice.  Now, with that in place, we have that 
$$
1 - \frac{3}{2x^{2}} \geq \frac{5}{8}
$$
and
$$
3 + \frac{7}{2x} \leq \frac{19}{4}.
$$
Overall then, we have that 
$$
\left|\frac{2+3x+x^{2}}{2x^{2}-3}-\frac{1}{2} \right| \leq \frac{19}{5|x|} \lt \epsilon
$$
Thus, letting $M = \frac{19}{5\epsilon}$, we have 
$$
\forall \epsilon \gt 0,~ \exists M = \frac{19}{5\epsilon}, |x|\gt M \rightarrow \left|\frac{2+3x+x^{2}}{2x^{2}-3}-\frac{1}{2} \right| \lt \epsilon.
$$

**Problem 2**: Prove that $f(x) = |x|$ is continuous for all $x\in\mathbb{R}$.

_Provide your answer here_ :

Using the triangle inequality, we have that 
$$
\left|\left|x\right|-\left|x_{0}\right| \right| \leq \left|x-x_{0} \right| \lt \epsilon.
$$
So letting $\delta = \epsilon$ we have that 
$$
\forall \epsilon \gt 0, \exists \delta = \epsilon, |x-x_{0}|\lt \delta \rightarrow \left|\left|x\right|-\left|x_{0}\right| \right| \lt \epsilon.
$$

**Problem 3**: Prove that $f(x)=\tan(x)$ is continuous for $x\in[0,\pi/2)$.  

_Provide your answer here_ : 

Were I you, letting $x_{0}\in[0,\pi/2)$, I would start this as
\begin{align}
\left|\tan(x) -\tan(x_{0})\right| = & \left|\frac{\sin(x)}{\cos(x)} -\frac{\sin(x_{0})}{\cos(x_{0})} \right| \\
= & \left|\frac{\sin(x)\cos(x_{0})-\cos(x)\sin(x_{0})}{\cos(x)\cos(x_{0})} \right|\\
= & \left|\frac{\sin(x-x_{0})}{\cos(x)\cos(x_{0})} \right|\\
= & \left|\frac{\sin(x-x_{0})}{\left(\cos(x)-\cos(x_{0})+\cos(x_{0})\right)\cos(x_{0})} \right|\\
\leq & \frac{\left|\sin(x-x_{0})\right|}{\left|\cos(x_{0})\right|}\frac{1}{\left|\left|\cos(x)-\cos(x_{0})\right|-\left|\cos(x_{0})\right|\right|}
\end{align}

From lecture, we know that 
$$
\lim_{x\rightarrow x_{0}} \cos(x) = \cos(x_{0}).
$$
We can then find a $\tilde{\delta}\gt 0$ such that 
$$
\left|\cos(x)-\cos(x_{0})\right| \lt \frac{|\cos(x_{0})|}{2},
$$
if $|x-x_{0}|\lt \tilde{\delta}$.  

Then, we see that if $|x-x_{0}|\lt \tilde{\delta}$, we find that 
$$
\left|\tan(x) -\tan(x_{0})\right| \lt \frac{2\left|\sin(x-x_{0})\right|}{\cos^{2}(x_{0})}.
$$
From lecture, we showed that  
$$
\left|\sin\left(x-x_{0}\right)\right| \leq \left|x-x_{0} \right|,
$$
and so we see finally that if we choose 
$$
\delta = \min\left\{\tilde{\delta},\frac{\epsilon \cos^{2}(x_{0})}{2}\right\},
$$
we have 
$$
\forall \epsilon \gt 0, \exists \delta = \min\left\{\tilde{\delta},\frac{\epsilon \cos^{2}(x_{0})}{2}\right\}, |x-x_{0}|\lt \delta \rightarrow \left|\tan(x) -\tan(x_{0})\right|\lt \epsilon.
$$

**Problem 4**: Prove that $\lim_{x\rightarrow\pi/2^{-}}\tan(x)=\infty$.  

_Provide your answer here_ :

So the easiest way to do this is to use the identities
$$
\tan(x) = \frac{\sin(x)}{\cos(x)} = \frac{\sin(x-\pi/2+\pi/2)}{\cos(x-\pi/2+\pi/2)} = \frac{\cos(\pi/2-x)}{\sin(\pi/2-x)}.
$$
Then, letting $u=\pi/2-x$, we can recast our problem as 
$$
\lim_{u\rightarrow 0^{+}}\frac{\cos(u)}{\sin(u)} = \infty.
$$
Moreover, from lecture, we showed that 
$$
\cos(u) \geq 1 - \frac{u^{2}}{2}
$$
and 
$$
\sin(u) \leq u,
$$
so we see that 
$$
\frac{\cos(u)}{\sin(u)}\geq \frac{1}{u}\left(1-\frac{u^{2}}{2}\right) 
$$

To make our lives simpler, we choose $0\lt u \lt 1$, so that $1-u^{2}/2 \gt 1/2$.  Therefore, we finally have that 
$$
\forall M\gt 0, \exists \delta = \min\left\{1,\frac{1}{2M}\right\}, |u|\lt \delta \rightarrow \left|\tan(x) \right|\gt M.
$$

**Problem 5**: Show that if $f(x)$ and $g(x)$ are continuous at a point $x_{0}\in\mathbb{R}$, then 
+ $f(x)+g(x)$ is continuous at $x_{0}$
+ $f(x)g(x)$ is continuous at $x_{0}$
+ If $g(x_{0})\neq 0$, $f(x)/g(x)$ is continuous at $x_{0}$.

_Provide your answer here_ :
#### 5a:
So for the first one, we see that 
$$
\left|(f(x)+g(x))-(f(x_{0})+g(x_{0}))\right| = \left|f(x)-f(x_{0}) + g(x)-g(x_{0}) \right|\leq \left|f(x)-f(x_{0})\right| + \left|g(x)-g(x_{0})\right|
$$
Then, there must be a $\delta_{1}$ such that if $|x-x_{0}|\lt \delta_{1}$ then 
$$
\left|f(x)-f(x_{0})\right|\lt \frac{\epsilon}{2}.
$$
Likewise, there must be a $\delta_{2}$ such that if $|x-x_{0}|\lt \delta_{2}$ then 
$$
\left|g(x)-g(x_{0})\right|\lt \frac{\epsilon}{2}.
$$
Therefore, we have shown 
$$
\forall \epsilon \gt 0 \exists \delta = \min\left\{\delta_{1},\delta_{2}\right\}, |x-x_{0}|\lt \delta \rightarrow \left|(f(x)+g(x))-(f(x_{0})+g(x_{0}))\right|\lt \epsilon
$$

#### 5b:
Proceeding in a similar fashion, we next look at 
\begin{align*}
\left|f(x)g(x)-f(x_{0})g(x_{0})\right| = & ~\left|(f(x)-f(x_{0}))g(x) + f(x_{0})(g(x)-g(x_{0}))\right|\\
= & ~\left|(f(x)-f(x_{0}))g(x)  + f(x_{0})(g(x)-g(x_{0}))\right|\\
\leq & ~\left|f(x)-f(x_{0})\right|\left|g(x)\right|  + \left|f(x_{0})\right|\left|g(x)-g(x_{0})\right|.
\end{align*}
As above, there must exist a $\delta_{1}$ such that $|x-x_{0}|\lt \delta_{1}$ implies that
$$
\left|g(x)-g(x_{0})\right|\lt \left|g(x_{0})\right|
$$
Thus, in turn, we have that if $\left|x-x_{0}\right|\lt\delta_{1}$ then 
\begin{align*}
\left|g(x)\right| = & ~\left|g(x)-g(x_{0})+g(x_{0})\right|\\
\leq & ~\left|g(x)-g(x_{0})\right| + \left|g(x_{0})\right|\\
\lt & ~2\left|g(x_{0})\right|,
\end{align*}
and therefore, we have that 
$$
\left|f(x)g(x)-f(x_{0})g(x_{0})\right| \leq 2\left|g(x_{0})\right|\left|f(x)-f(x_{0})\right|  + \left|f(x_{0})\right|\left|g(x)-g(x_{0})\right|
$$

Finally, there exists $\delta_{2}$ and $\delta_{3}$ such that $|x-x_{0}|\lt \delta_{2}$ implies
$$
\left|f(x)-f(x_{0})\right|\lt \frac{\epsilon}{4|g(x_{0})|}.
$$
and $|x-x_{0}|\lt \delta_{3}$ implies
$$
\left|g(x)-g(x_{0})\right|\lt \frac{\epsilon}{2|f(x_{0})|}.
$$

Therefore, if $|x-x_{0}|\lt\tilde{\delta}=\min\left\{\delta_{1},\delta_{2},\delta_{3}\right\}$, then we have that 
$$
\left|f(x)g(x)-f(x_{0})g(x_{0})\right| \leq \epsilon.
$$

#### 5c:
We focus on proving that 
$$
\lim_{x\rightarrow x_{0}}\frac{1}{g(x)} = \frac{1}{g(x_{0})}
$$
and then use the result from 5b to complete the proof.  So, we start with 
$$
\left|\frac{1}{g(x)} - \frac{1}{g(x_{0})} \right| = \frac{\left|g(x)-g(x_{0})\right|}{\left|g(x)\right|\left|g(x_{0})\right|}
$$
From continuity of $g(x)$, we know there must be some $\delta_{1}$ such that if $|x-x_{0}|\lt \delta_{1}$, then 
$$
\left|g(x)-g(x_{0})\right|\lt \frac{|g(x_{0})|}{2}.
$$
Therefore, if $|x-x_{0}|\lt \delta_{1}$, we have that 
\begin{align*}
\left|g(x)\right| = & \left|g(x)-g(x_{0}) + g(x_{0})\right|\\
\geq & \left|\left|g(x_{0})\right|-\left|g(x)-g(x_{0})\right| \right|\\
\gt & \frac{|g(x_{0})|}{2}.
\end{align*}

We have now shown that if $|x-x_{0}|\lt \delta_{1}$, then 
$$
\left|\frac{1}{g(x)} - \frac{1}{g(x_{0})} \right| = \frac{2\left|g(x)-g(x_{0})\right|}{g^{2}(x_{0})}
$$
We likewise know that there is a $\delta_{2}$ such that if $|x-x_{0}|\lt \delta_{2}$, then 
$$
\left|g(x)-g(x_{0})\right|\lt \frac{g^{2}(x_{0})\epsilon}{2}.
$$
Thus, we have now shown that 
$$
\forall \epsilon \gt 0, \exists \delta=\min\left\{\delta_{1},\delta_{2}\right\}, |x-x_{0}|\lt \delta \rightarrow \left|\frac{1}{g(x)} - \frac{1}{g(x_{0})}\right| \lt \epsilon.
$$

**Problem 6**: Using the previous problem, show that any polynomial of the form $\pi(x) =\sum_{m=0}^{n}a_{m}x^{m}$ is continuous for all $x\in\mathbb{R}$.

_Provide your answer here_ :

So we just use the results from above and quick inductive arguments.  So, clearly any constant function, say $f(x)=c$ is continuous.  Similarly, so is $f(x)=x$.  We see that since 
$$
f(x) = cx^{n} = cxx^{n-1},
$$
then if $x^{n-1}$ is continuous, then $cx^{n}$ is since products of continuous functions are continuous.  Likewise then, noting that 
$$
\pi(x) = a_{n}x^{n} + \sum_{m=0}^{n-1}a_{m}x^{m}
$$
then, by induction, if $\sum_{m=0}^{n-1}a_{m}x^{m}$ is continuous, then $\pi(x)$ must be as well since the sum of continuous functions are continuous.  

**Problem 7**: Let $f:\mathbb{R}\rightarrow \mathbb{R}$ be a function such that $f(x+y)=f(x)+f(y)$.  Assume that $\lim_{x\rightarrow 0}f(x)=L$ exists.  
+ Prove that $L=0$.  _Hint_ : use $f(x+x)=f(x)+f(x)=2f(x)$.    
+ Prove that $f(x)$ is continuous for all $x\in \mathbb{R}$.  _Hint_ : use $f(x)=f(x-x_{0}+x_{0})=f(x-x_{0})+f(x_{0})$.    

_Provide your answer here_ :  

So a great way to start the first problem is to note that 
$$
f(x) = f\left(\frac{x}{2} + \frac{x}{2}\right) = 2f\left(\frac{x}{2}\right)
$$
Thus, choosing the sequence $x_{n}=1/2^{n}$, $n\geq 1$, then we see that $x_{n+1}=x_{n}/2$, and thus 
$$
f(x_{n+1}) = f\left(\frac{x_{n}}{2}\right) = \frac{1}{2}f(x_{n}).  
$$
Therefore we see that
$$
f(x_{n+1}) = \frac{1}{2^{n}}f(x_{1}), ~ n\geq 1.  
$$
Using THM 4.1.1 from the book will take care of the rest since clearly
$$
\lim_{n\rightarrow\infty}f(x_{n+1}) = 0,
$$
and 
$$
\lim_{n\rightarrow\infty} x_{n+1} = 0.
$$
So if $\lim_{x\rightarrow 0}f(x)=L$ by THM 4.1.1., we see that $L=0$.  From the identity 
$$
f(x) = f\left(\frac{x}{2}\right),
$$
we see $f(0)=2f(0)$, and therefore $f(0)=0$.  Thus 
$$
\lim_{x\rightarrow 0}f(x) = 0 = f(0),
$$
and $f(x)$ is continuous at $x=0$.

As for the next part of the problem, from the hint we see that 
$$
\left|f(x)-f(x_{0})\right| = \left|f(x-x_{0})\right|.
$$
From the first problem, we see that for a given $\epsilon\gt 0$, $\exists \delta\gt 0$ such that is  $|x-x_{0}|\lt \delta$ then 
$$
\left|f(x-x_{0})\right| \lt \epsilon.  
$$
Therefore, we see 
$$
\forall \epsilon \gt 0,~ \exists \delta \gt 0, ~ |x-x_{0}|\lt \delta \rightarrow \left|f(x)-f(x_{0})\right| \lt \epsilon.
$$
